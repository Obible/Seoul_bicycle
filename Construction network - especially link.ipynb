```python
df1 = pd.read_csv('bike(2020.01).csv',\
                   encoding='cp949').drop(['자전거번호', '대여일시','대여 대여소명', '대여거치대', '반납일시',\
                                           '반납대여소명', '반납거치대', '이용시간','이용거리'],axis=1)
df11 = pd.read_csv('bike(2020.02).csv',\
                   encoding='cp949').drop(['자전거번호', '대여일시','대여 대여소명', '대여거치대', '반납일시',\
                                           '반납대여소명', '반납거치대', '이용시간','이용거리'],axis=1)
df111 = pd.read_csv('bike(2020.03).csv',\
                   encoding='cp949').drop(['자전거번호', '대여일시','대여 대여소명', '대여거치대', '반납일시',\
                                           '반납대여소명', '반납거치대', '이용시간','이용거리'],axis=1)
df1111 = pd.read_csv('bike(2020.04).csv',\
                   encoding='cp949').drop(['자전거번호', '대여일시','대여 대여소명', '대여거치대', '반납일시',\
                                           '반납대여소명', '반납거치대', '이용시간','이용거리'],axis=1)
```

```python
'''

3, 4월 데이터가 반납거치소번호에서 00454, 454 둘 다 나옴.
dataframe copy -> NaN 나오는 거 찾아서 원래 dataframe 값 받기.

'''

c111 = df111.copy()
c1111 = df1111.copy()

df1['반납대여소번호'] = df1['반납대여소번호'].str.lstrip("0")
df11['반납대여소번호'] = df11['반납대여소번호'].str.lstrip("0")
df111['반납대여소번호'] = df111['반납대여소번호'].str.lstrip("0")
df1111['반납대여소번호'] = df1111['반납대여소번호'].str.lstrip("0")

# NaN 나옴.
df1111
```

```python
# dff = df11111.copy()

df67 = c1111[df1111['반납대여소번호'].isnull()] # df1111의 NaN을 찾아서 c111로 반환한 dataframe을 새로운 df에 저장.
df1111[df1111['반납대여소번호'].isnull()] = df67 # 원래의 df1111이 NaN을 갖는 행에 새로운 df의 값을 할당.

df1111[df1111['반납대여소번호'].isnull()] # NaN 값 나오는지 확인.
```

```python
df67 = c111[df111['반납대여소번호'].isnull()] # df1111의 NaN(->True)을 찾아서 c111로 반환한 dataframe을 새로운 df에 저장.
df111[df111['반납대여소번호'].isnull()] = df67 # 원래의 df1111이 NaN을 갖는 행에 새로운 df의 값을 할당.

df111[df111['반납대여소번호'].isnull()] # NaN 값 나오는지 확인.
```

```python
bike_filtered = bike.drop(['자치구', '상세주소',\
                           '설치\n시기', 'LCD', 'QR', 'LCD + QR', '운영\n방식'],axis=1)
bike_filtered
```

```python
'''

문제: dataframe에 '반납대여소번호' 열에 숫자가 아니라 한글이 있음.

1. 해당 column에서 한글인 row만 저장해서
2. station dataframe에서 '보관소(대여소)명'에서 같은 글자 찾고
3. 그 때의 'station number'를 찾는다.
4. list에 하나씩 append하고 원래 dataframe에 값 할당해주기.

+ 모든 string int64로 바꾸기

'''

find_str = df3[df3['반납대여소번호'].str.isdigit()==False]
x = find_str['반납대여소번호']

b_id = bike_filtered['station number']
b_name = bike_filtered['보관소(대여소)명']

newi = []

for i in x:
    ff = str(b_id[b_name==i])
    f = ff.split()
    
    newi.append(f[1])

print(len(newi), len(x))
print(newi)
```

```python
df3.loc[df3['반납대여소번호'].str.isdigit()==False, '반납대여소번호'] = newi

# 숫자가 아닌 글자가 있는지 확인
df3[df3['반납대여소번호'].str.isdigit()==False]
```

```python
# '반납대여소번호' 열을 integer로 바꾸기.
df3['반납대여소번호'] = df3['반납대여소번호'].astype(str).astype(int)
print(df3.dtypes)
```

```python
# 중복이 나와서 한 번 더 중복 값을 합산해줌.
df3 = df3.groupby(['대여 대여소번호', '반납대여소번호'], as_index=False).sum() 
df3[df3.duplicated(['대여 대여소번호', '반납대여소번호'])]
```
